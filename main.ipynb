```python
import pandas as pd

from pprint import pprint

import matplotlib.pyplot as plt

import seaborn as sns

import plotly.express as px

import ast
```

```python
# Importing data-
data= pd.read_json('data.json',  typ = 'series')
```

```python
data.head()
```

```python
data['grants']
```

```python
grant_str = str(data['grants'])
```

```python
grant_str = grant_str[1:]
```

```python
grant_str = grant_str[:-1]
```

```python
pprint(grant_str)
```

```python
av = ast.literal_eval(grant_str)
```

```python
title=[]
tags=[]
total_amount_received_usd_life=[]
estimated_round_3_clr_match_usd=[]
url = []
admin_profile_name = []
admin_address = []
no_of_contributions = []

for i in range(len(av)):
    title.append(av[i]['title'])
    tags.append(av[i]['tags'])
    total_amount_received_usd_life.append(av[i]['total_amount_received_usd_life'])
    estimated_round_3_clr_match_usd.append(av[i]['estimated_round_3_clr_match_usd'])
    url.append(av[i]['url'])
    admin_profile_name.append(av[i]['admin_profile_name'])
    admin_address.append(av[i]['admin_address'])
    no_of_contributions.append(len(av[i]['contributions']))
```

```python
len(av[0]['contributions'])
```

```python
data = pd.DataFrame({'title':title,
                     'tags':tags,
                     'total_amount_received_usd_life':total_amount_received_usd_life,
                     'estimated_round_3_clr_match_usd':estimated_round_3_clr_match_usd,
                     'url': url,
                     'admin_profile_name':admin_profile_name,
                     'admin_address':admin_address,
                     'no_of_contributions':no_of_contributions})
```
